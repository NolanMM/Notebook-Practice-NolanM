## III. Numerical Data Ploting

### 1.relplot()

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style='time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette='ch:r=-0.85, l= 0.95')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette='ch:r=-0.85, l= 0.95',sizes ='size')
```

### Note: 
#### - Hue: Changing the color
#### - Style: Changing the label
#### - Size: Changing the size(dimensions) of element plot

#### Custom size

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette='ch:r=-0.85, l= 0.95', size = 'size', sizes = (15,200))
```

### 2.Line Plot

```python
df = pd.DataFrame(dict(time=np.arange(500),value = randn(500).cumsum()))
```

```python
df
```

```python
sns.relplot(x='time', y ='value', kind ='line', data=df, sort = True) # If Data is messy we can using sort = True
```

```python
df_draf_no_sort = pd.DataFrame(randn(500,2).cumsum(axis=0), columns = ['time','value'])
```

```python
sns.relplot(x='time', y ='value', kind ='line', data=df_draf_no_sort, sort = False)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.relplot(x='timepoint', y ='signal', kind ='line', data=fmri) # by default ci = True # And CI = Confident Interval
                                                                 # At a x point, the y value can have value in range of Interval
```

```python
sns.relplot(x='timepoint', y ='signal', kind ='line', data=fmri, ci=False) # ci = False
```

```python
sns.relplot(x='timepoint', y ='signal', kind ='line', data=fmri, ci='sd') # ci = sd (means Confident Intervval = Standard Deviation)
```

```python
sns.relplot(x='timepoint', y ='signal', estimator=None, kind ='line', data=fmri) # Plot real dataset
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='event', kind ='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='region', style='event', kind ='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='region', style='event', kind ='line', data=fmri, markers= True)
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='region', style='event', kind ='line', data=fmri, markers= True, dashes=False)
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='event', style='event', kind ='line', data=fmri, markers= True)
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='event', style='event', kind ='line', data=fmri, markers= True, dashes = False)
```

#### Multiple Measurements

```python
sns.relplot(x='timepoint', y ='signal', hue ='region', units ='subject', estimator = None, kind = 'line', data=fmri)
```

```python
sns.relplot(x='timepoint', y ='signal', hue ='region', units ='subject', estimator = None, kind = 'line', data=fmri.query("event == 'stim'"))
```

#### Style Sematics

```python
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```python
sns.relplot(x='time', y = 'firing_rate', data = dots,  kind = 'line')
```

```python
sns.relplot(x='time', y = 'firing_rate', data = dots,  kind = 'line', hue = 'coherence')
```

```python
sns.relplot(x='time', y = 'firing_rate', data = dots,  kind = 'line', hue = 'coherence', style = 'choice')
```

```python
palette = sns.cubehelix_palette(light = 0.5, n_colors=6) # Darker
sns.relplot(x='time', y = 'firing_rate', data = dots,  kind = 'line', hue = 'coherence', style = 'choice',  palette=palette)
```

#### Custom Line Width

```python
sns.relplot(x='time', y = 'firing_rate', data = dots,  kind = 'line', size='coherence', style='choice', hue='coherence')
```
